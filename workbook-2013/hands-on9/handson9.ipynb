```python
import numpy as np
```

```python
iris = np.loadtxt('iris.txt',skiprows=1,usecols=range(4))
# note we used only the first 4 cols, the last column gives species names
```

```python
iris.shape
```

```python
import scipy.spatial.distance as dist
d = dist.pdist(iris, 'euclidean')
```

```python
d.shape
```

```python
(150*149)/2  # 
```

```python
import scipy.cluster.hierarchy as hier
```

```python
ilink = hier.linkage(d)
```

```python
dendro = hier.dendrogram(ilink)
```

```python
dendro = hier.dendrogram(ilink, color_threshold=0.5) 
# colors the subtrees at a different threshold
```

```python
species = np.loadtxt('iris.txt', dtype=str, skiprows=1, usecols=[4]) # get species names
```

```python
dendro = hier.dendrogram(ilink, color_threshold=0.6,labels=species) 
```

```python
d2 = dist.pdist(iris, 'cityblock')
```

```python
iupgma = hier.average(d2)
```

```python
dendro2 = hier.dendrogram(iupgma, labels=species, orientation='right', leaf_font_size=10)
```
